```python
# tight_layout()    #  
x=np.linspace(0,5,10)
y = x**3
fig, axes = plt.subplots(nrows=1,ncols=4)
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

plt.tight_layout()
plt.show()
```

```python
#legend()    #
x=np.linspace(0,5,10)
y = x**3
fig, axes = plt.subplots(nrows=1,ncols=4)
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    ax.legend(['A simple line'])

plt.tight_layout()
plt.show()
```

```python
#   2 y
import numpy as np
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()
t = np.arange(0.01, 10.0, 0.01)
s1 = np.exp(t)
ax1.plot(t, s1, 'b-')
ax1.set_xlabel('time $(s^2)$')
ax1.set_ylabel('exp', color='b')
ax2 = ax1.twinx()    #ax1 x  ax2 
s2 = np.sin(2*np.pi*t)
ax2.plot(t, s2, 'r.')
ax2.set_ylabel('sin', color='r')
plt.show()
```

```python
# fig.add_axis()    # 
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])    #left, bottom, width, height
t = np.arange(0.01, 10.0, 0.01)
s = np.sin(2*np.pi*t)
ax.plot(t, s, 'b-')
ax.set_xlabel('time (s)')
ax.set_ylabel('exp')
ax.set_title('About as simple as it gets, folks')

#insert
ax2 = fig.add_axes([0.15, 0.15, 0.3, 0.3])
ax2.plot(t, s, 'r-')
ax2.set_xlabel('time (s)')
ax2.set_ylabel('sin')
ax2.set_title('insert')

plt.show()
```

```python
#stacked bar chart
import matplotlib.pyplot as plt
import numpy as np
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
p1 = plt.bar(ind, menMeans, width)
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans)
plt.ylabel('Scores')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.title('haha')

plt.show()
```

```python
#stacked barh + center align
import matplotlib.pyplot as plt
import numpy as np
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
p1 = plt.barh(ind, menMeans, width,align='center')
p2 = plt.barh(ind, womenMeans, width, left=menMeans,align='center')
plt.ylabel('Scores')
plt.yticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.xticks(np.arange(0, 81, 10))
plt.title('haha')


plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```

```python
#scatter with colorbar
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.random.randn(100)
y = np.random.randn(100)
c = np.random.randn(100)
s = np.random.randn(100)*100
im = ax.scatter(x,y,c=c,s=s,cmap=plt.cm.jet)
fig.colorbar(im,ax=ax)

im.set_clim([0.,1.])

plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
import numpy as np

labels = 'python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
explode = (0.1, 0, 0, 0)    #explode 1st slice

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
shadow=True, startangle=90)
ax.axis('equal')    # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
